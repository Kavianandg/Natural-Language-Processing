<h1 style="color:red;" style="font-family:georgia;"><i>Cosine Similarity
    </i></h1>

```python
data_1 = "Data is the oil of the digital economy"
data_2 = "Data is a new oil"

data = [data_1, data_2]
```

## Using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer()
vector_matrix = count_vect.fit_transform(data)
print(vector_matrix)
```

```python
tokens = count_vect.get_feature_names()
print(tokens)
```

```python
vocab = count_vect.vocabulary_ 
vocab
```

```python
vec_data_1 = count_vect.transform([data_1]).toarray()
print(vec_data_1)
```

```python
vec_data_2 = count_vect.transform([data_2]).toarray()
print(vec_data_2)
```

```python
matrix = vector_matrix.toarray()
print(matrix)
```

```python
import pandas as pd

def create_dataframe(matrix, tokens):

    doc_names = [f'doc_{i+1}' for i, _ in enumerate(matrix)]
    df = pd.DataFrame(data=matrix, index=doc_names, columns=tokens)
    return(df)
```

```python
create_dataframe(matrix,tokens)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity_matrix = cosine_similarity(vector_matrix)
create_dataframe(cosine_similarity_matrix,['doc_1','doc_2'])
```

```python
print("Cosine Similarity = ",(cosine_similarity(vec_data_1,vec_data_2))[0][0])
```

## Using TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

Tfidf_vect = TfidfVectorizer()
vector_matrix = Tfidf_vect.fit_transform(data)

tokens = Tfidf_vect.get_feature_names()
create_dataframe(vector_matrix.toarray(),tokens)
```

```python
cosine_similarity_matrix = cosine_similarity(vector_matrix)
create_dataframe(cosine_similarity_matrix,['doc_1','doc_2'])
```

```python
vec_data_1 = Tfidf_vect.transform([data_1]).toarray()
print(vec_data_1)
```

```python
vec_data_2 = Tfidf_vect.transform([data_2]).toarray()
print(vec_data_2)
```

```python
print("Cosine Similarity = ",(cosine_similarity(vec_data_1,vec_data_2))[0][0])
```

<h1 style="color:blue;" style="font-family:georgia;"><i>Simple Text Classifier    </i></h1>

```python
from nltk.corpus import names
import random
```

```python
male_name =[(name, 'male') for name in names.words('male.txt')]
female_name = [(name, 'female') for name in names.words('female.txt')]
```

```python
print(male_name,female_name)
```

```python
labelled_name = male_name + female_name
random.shuffle(labelled_name)
```

```python
print(labelled_name)
```

```python
print(len(labelled_name))
```

```python
def gender_features(word): #gives last letter of the word
    return {'last_letter':word[-1]}
```

```python
featuresets = [(gender_features(n),gender) for (n,gender) in labelled_name] 
```

```python
featuresets
```

```python
train_set, test_set = featuresets[500:], featuresets[:500]
```

```python
print(len(train_set),len(test_set))
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
train_set_acc = nltk.classify.accuracy(classifier, train_set)
test_set_acc = nltk.classify.accuracy(classifier, test_set)
```

```python
print("Accuracy on Train dataset = ", train_set_acc)
print("Accuracy on Test dataset = ", test_set_acc)
```

```python
classifier.classify(gender_features("Kavianand"))
```

```python
classifier.classify(gender_features("Kavi"))
```
